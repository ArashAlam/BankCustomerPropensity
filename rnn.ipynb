# Recurrent Neural Network

## Part 1 - Data Preprocessing

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

### Importing the libraries

```
datasettrain = pd.read_csv('NVDA_train.csv')
trainsetX=datasettrain.iloc[:,7:13].values
trainsetY1=datasettrain.iloc[:,13:14].values
trainsetY2=datasettrain.iloc[:,14:15].values


datasettest = pd.read_csv('NVDA_test.csv')
testsetX=datasettest.iloc[:,7:13].values
testsetY1=datasettest.iloc[:,13:14].values
testsetY2=datasettest.iloc[:,14:15].values

```

### Feature Scaling

```
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
xtrain = sc.fit_transform(trainsetX)
xtest = sc.transform(testsetX)
ytrain=trainsetY2
ytest= testsetY2

ytrain
```

```
annn = tf.keras.models.Sequential()
annn.add(tf.keras.layers.Dense(units=5,activation="relu"))
annn.add(tf.keras.layers.Dense(units=8,activation="relu"))
annn.add(tf.keras.layers.Dense(units=5,activation="relu"))

annn.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))


annn.compile(optimizer='adam', loss = 'binary_crossentropy' , metrics =['accuracy'] )


annn.fit(xtrain, ytrain,batch_size=200 , epochs=1000)
```

### Creating a data structure with 60 timesteps and 1 output

```
result=annn.predict(sc.transform(xtest))


plt.plot(annn.predict(sc.transform(xtest)))
plt.plot(ytest)
```
